```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Ações_puras.xlsx",sheet_name="Principal")
df_principal.head()
```

```python
df_total_acoes = pd.read_excel("/content/Ações_puras.xlsx",sheet_name="Total_de_acoes")
df_total_acoes = df_total_acoes.rename(columns={'Código':'codigo','Qtde. Teórica':'qtde_teorica'}).copy()
df_total_acoes.head()
```

```python
df_ticker = pd.read_excel("/content/Ações_puras.xlsx",sheet_name="Ticker")
df_ticker = df_ticker.rename(columns={'Ticker':'ticker','Nome':'nome'}).copy()
df_ticker.head()
```

```python
df_chatgpt = pd.read_excel("/content/Ações_puras.xlsx",sheet_name="chatgpt")
df_chatgpt = df_chatgpt.rename(columns={'Nome da Empresa':'nome_empresa','Segmento':'segmento','Idade (em anos)':'idade'}).copy()
df_chatgpt.head()
```

```python
df_principal = df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo','Data':'data','Último (R$)':'valor_final','Var. Dia (%)':'vlr_dia_pct'}).copy()
df_principal
```

```python
df_principal['vlr_pct'] = df_principal['vlr_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['vlr_pct']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes,left_on='ativo',right_on='codigo', how='left')
df_principal = df_principal.drop(columns=['codigo'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) *  df_principal['qtde_teorica']
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker,left_on='ativo',right_on='ticker', how='left')
df_principal = df_principal.drop(columns=['ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt,left_on='nome',right_on='nome_empresa', how='left')
df_principal = df_principal.drop(columns=['nome_empresa'])
df_principal
```

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Menor que 50 anos" if x < 50 else "Entre 50 e 100"))
df_principal
```

```python
# Função para calcular o maior valor
def calcular_maior(df_principal):
    return df_principal.loc[df_principal['variacao_rs'].idxmax()]

# Função para calcular o menor valor
def calcular_menor(df_principal):
    return df_principal.loc[df_principal['variacao_rs'].idxmin()]

# Função para calcular a média
def calcular_media(df_principal):
    return df_principal['variacao_rs'].mean()

# Função para calcular a média dos que subiram
def calcular_media_subiram(df_principal):
    return df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()

# Função para calcular a média dos que desceram
def calcular_media_desceram(ddf_principalf):
    return df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

# Aplicando as funções
maior_valor = calcular_maior(df_principal)
menor_valor = calcular_menor(df_principal)
media = calcular_media(df_principal)
media_subiram = calcular_media_subiram(df_principal)
media_desceram = calcular_media_desceram(df_principal)

# Resultados
print("Maior valor:")
print(maior_valor)
print("\nMenor valor:")
print(menor_valor)
print("\nMédia:")
print(media)
print("\nMédia dos que subiram:")
print(media_subiram)
print("\nMédia dos que desceram:")
print(media_desceram)
```
